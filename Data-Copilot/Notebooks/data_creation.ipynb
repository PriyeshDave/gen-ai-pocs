```python
def assign_manufacturer(laptop_model):
  if laptop_model.__contains__('macbook'):
    return 'Apple'
  elif laptop_model == 'dell':
    return 'DELL'
  elif laptop_model == 'hp':
    return 'HP'
  else:
    return 'Lenovo'
```

```python
asset_df['manufacturer'] = asset_df['laptop_model'].apply(assign_manufacturer)
asset_df.head()
```
